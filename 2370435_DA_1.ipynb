```python
missing_values = ["Na", "na"]
df = pd.read_csv("test_data.csv", na_values = missing_values)
df
```

```python
# Filling missing values in the 'Math_Score' column with the mean
df['Math_Score'] = df['Math_Score'].fillna(df['Math_Score'].mean())

# Filling missing values in the 'Math_Score' column with the median
df['Math_Score'] = df['Math_Score'].fillna(df['Math_Score'].median())

# Filling missing values in the 'Math_Score' column with the standard deviation
df['Math_Score'] = df['Math_Score'].fillna(df['Math_Score'].std())

# Filling missing values in the 'Math_Score' column with the minimum value
df['Math_Score'] = df['Math_Score'].fillna(df['Math_Score'].min())

# Filling missing values in the 'Math_Score' column with the maximum value
df['Math_Score'] = df['Math_Score'].fillna(df['Math_Score'].max())
```

```python
import numpy as np
from scipy import stats

z = np.abs(stats.zscore(df['Math_Score']))
print(z)
```

```python
sorted_rscore=sorted(df['Reading_Score'])
sorted_rscore
```

```python
q1=np.percentile(sorted_rscore,25)
q3=np.percentile(sorted_rscore,75)
print(q1,q3)
```

```python
iqr=q3-q1
```

```python
lwr_bound=q1-(1.5*iqr)
upr_bound=q3+(1.5*iqr)
print(lwr_bound,upr_bound)
```

```python
r_outliers=[]
for x in sorted_rscore:
  if((x<lwr_bound) or (x>upr_bound)):
    r_outliers.append(x)
print(r_outliers)
```

```python
new_df=df
for i in r_outliers:
  new_df=new_df.drop(i,inplace=True)
new_df
```
