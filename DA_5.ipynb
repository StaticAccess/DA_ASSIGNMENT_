**The Joint PLot**


```python
#jointplot() will show distribution of one column wrt to other
#accepts 3 parameter 1. xaxis column y-axis column and data "kind" is used for type of plot
sns.jointplot(x="age",y="fare",data=data)
```

```python
sns.jointplot(x="age",y="fare",data=data,kind="hex") #for hexagon kind=hex
```

```python
import seaborn as sns
import matplotlib.pyplot as plt # Ensure matplotlib.pyplot is imported

ax = sns.countplot(df["Survived"])
ax.set_xticklabels(["Not Survived", "Survived"])
plt.show() # Call show from the matplotlib.pyplot module
```

```python
import matplotlib.pyplot as plt


survived_counts = df['Survived'].value_counts()

# Create the pie chart
plt.pie(survived_counts, labels=survived_counts.index, autopct="%.2f")
plt.title('Survival Rate')

# Display the chart
plt.show()
```
